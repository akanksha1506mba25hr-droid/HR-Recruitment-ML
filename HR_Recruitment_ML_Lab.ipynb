```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
# Fix random values so output is same every time
np.random.seed(42)

n = 50

data = {
    'experience_years': np.random.randint(0, 11, n),
    'test_score': np.random.randint(40, 100, n),
    'interview_score': np.random.randint(40, 100, n),
    'skill_match': np.random.randint(1, 4, n)
}

df = pd.DataFrame(data)

df.head()
```

```python
df['selected'] = (
    ((df['experience_years'] >= 3) &
     (df['test_score'] >= 60) &
     (df['interview_score'] >= 60)) |

    ((df['skill_match'] == 3) &
     (df['interview_score'] >= 70))
).astype(int)

df.head()
```

```python
X = df[['experience_years', 'test_score', 'interview_score', 'skill_match']]
y = df['selected']
```

```python
model = DecisionTreeClassifier(max_depth=4, random_state=42)
model.fit(X, y)
```

```python
sample = [[4, 75, 80, 3]]

prediction = model.predict(sample)

prediction
```

```python
if prediction[0] == 1:
    print("Candidate Likely to be Selected")
else:
    print("Candidate Likely to be Rejected")
```
